### Where?

Which airports are in the `Pacific/Port_Moresby` time zone? What city and country?

```python
pd.read_sql("""

SELECT name, city, country
FROM 
    airports
    WHERE 
        timezone == "Pacific/Port_Moresby"

""", con)
```

### Further down the rabbit hole!

Which countries have between 50 and 100 airports?

```python
pd.read_sql("""

SELECT COUNT() as num_airports, country
FROM 
    airports
GROUP BY
    country
    HAVING (num_airports >= 50 and num_airports <=100)
ORDER BY num_airports DESC

""", con)
```

### Joins!

What is the name of the airline has the most routes? The fewest?

```python
pd.read_sql("""

    SELECT 
        a.name
        , COUNT() as num_routes
    FROM airlines a
        LEFT JOIN routes r
            ON a.id = r.airline_id
    GROUP BY a.name
    ORDER By num_routes DESC

""", con)
```

```python
pd.read_sql("""

    SELECT 
        a.name
        , COUNT() as num_routes
    FROM airlines a
        LEFT JOIN routes r
            ON a.id = r.airline_id
    GROUP BY a.name
        HAVING num_routes == 1

""", con)
```

### Subqueries/CTEs!

What is the average number of routes per airline?

```python
subqry = """
    SELECT airline, COUNT() as num_routes
    FROM routes
    GROUP BY airline
    ORDER BY COUNT() DESC

"""
```

```python
qry = """
    WITH route_counts AS (SELECT airline, COUNT() as num_routes
    FROM routes
    GROUP BY airline
    ORDER BY COUNT() DESC)
    
    SELECT AVG(num_routes)
    FROM route_counts

"""
pd.read_sql(qry, con)
```
